# USB Trace Analysis

Use this notebook to inspect sanitized CSV exports or reviewed trace summaries from the `data/` folder.

Start with redacted data only, then promote any reusable parsing logic into `scripts/`.

```python
from pathlib import Path
import csv

data_dir = Path('../data')
csv_files = sorted(data_dir.glob('*.csv'))
print(f'Found {len(csv_files)} CSV file(s) in {data_dir.resolve()}')

for path in csv_files:
    with path.open(newline='', encoding='utf-8') as handle:
        row_count = sum(1 for _ in csv.reader(handle))
    print(f'{path.name}: {row_count} row(s)')
```
